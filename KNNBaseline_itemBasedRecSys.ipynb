```python
from surprise import KNNBaseline

# Initialize the algorithm
sim_options = {'name': 'cosine', 'user_based': False}  # Use cosine similarity for item-item
knn_model = KNNBaseline(sim_options=sim_options)

# Fit the model
knn_model.fit(trainset)

# Find similar items
item_id = trainset.to_raw_iid(1307)
print(item_id)
k = 10  # Number of similar items to retrieve
similar_items = knn_model.get_neighbors(trainset.to_inner_iid(item_id), k)

print("Similar items to item", item_id, ":")
for item in similar_items:
    print(trainset.to_raw_iid(item))
```

```python
import pickle

with open('KNNBaseline_itemBaseRec.pkl', 'wb') as f:
    # Step 4: Use pickle.dump()
    pickle.dump(knn_model, f)
```
